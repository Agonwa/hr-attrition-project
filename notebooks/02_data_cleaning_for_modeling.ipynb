```python
# Import essential libraries
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/HP/Desktop/HR ATTRITION PROJECT/data/WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Quick overview of the dataset
print("Original Shape:", df.shape)
df.head()
```

```python
# Convert attrition to binary 
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Confirm the conversion
df.head()
```

```python
# Check how many unique values each column has
unique_counts = df.nunique().sort_values()
unique_counts
```

EmployeeNumber is an Identifier and needs to be dropped

```python
# Identify constant columns
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
print("Constant Columns:", constant_cols)
```

```python
# Drop constant and irrelevant identifier columns
df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'], inplace=True)
```

```python
# Conform the dropping
df.head()
```

4 columns dropped already remaining 31 columns

```python
# Check for duplicates
print("Duplicate rows:", df.duplicated().sum())

# Check for missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Check columns that are still categorical 
cat_cols = df.select_dtypes(include='object').columns.tolist()
cat_cols
```

```python
# Map binary categorical variables explicitly
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})
```

```python
# One-hot encode multi-class categorical variables
df = pd.get_dummies(df, columns=[
    'BusinessTravel', 'Department', 'EducationField',
    'JobRole', 'MaritalStatus'
], drop_first=True)
```

```python
# Check the cleaned dataset
print("Cleaned Dataset Shape:", df.shape)
df.info()

# Preview cleaned data
df.head()
```

```python
# Save cleaned dataset for modeling and EDA
df.to_csv("../data/cleaned_hr_data.csv", index=False)

```
