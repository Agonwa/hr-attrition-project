```python
# Import essential libraries
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/HP/Desktop/HR ATTRITION PROJECT/data/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df.head()
```

```python
# Convert 'Attrition' from Yes/No to numeric (1 for Yes, 0 for No) for aggregation and visualization purposes
df["AttritionNumeric"] = df["Attrition"].map({"Yes": 1, "No": 0})
```

```python
# Create age group bins to categorize employees by age range
bins = [17, 25, 35, 45, 60]
labels = ["18-25", "26-35", "36-45", "46-60"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)
```

```python
# Create salary bands by dividing MonthlyIncome into 4 quantile-based groups
# Labels: Low, Medium, High, Very High
df["SalaryBand"] = pd.qcut(df["MonthlyIncome"], q=4, labels=["Low", "Medium", "High", "Very High"])
```
